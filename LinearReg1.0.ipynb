```python
#import lib
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# drop iq col
df.drop('IQ',axis=1,inplace=True)
df.head()
```

```python
#selecting input and output col
X=df.iloc[:,0]   #input data independent var
Y=df.iloc[:,1]   #label data dependent var
```

```python
#spliting data in train test data using sklearn
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(X,Y,test_size=0.2,random_state=0)
```

```python
#shape of each var
print(x_train.shape)
print(y_train.shape)
print(x_test.shape)
print(y_test.shape)
```

```python
# apply linear regression on Training data
from sklearn.linear_model import LinearRegression
LR=LinearRegression()
```

```python
x_train=x_train.values.reshape(-1,1)
```

```python
LR.fit(x_train,y_train)
```

```python
y_pred=LR.predict(x_test.values.reshape(-1,1))
```

```python
print(y_test,y_pred)
```

```python
#evaluation using sklearn metrics
from sklearn.metrics import mean_squared_error,r2_score
mean_squared_error(y_test,y_pred)
```

```python
#creating ML algo
class ML_LR:
  def __init__(self):
    self.beta =None #placeholder for Coefficients

  def fit(self, x_train, y_train):
    assert x_train.shape[0]>0

    x_train =np.insert(x_train,0,1,axis=1)
    self.beta =np.linalg.pinv(x_train.T @ x_train) @ x_train.T @ y_train

  def predict(self,x_test):
    if self.beta is None:
      raise ValueError("Model is not trained")

    x_test =np.insert(x_test,0,1,axis=1)
    return x_test @ self.beta
```

```python
lr=ML_LR()
```

```python
lr.fit(x_train,y_train)
```

```python
x_test=x_test.values.reshape(-1,1)
```

```python
lr.predict(x_test)
```
